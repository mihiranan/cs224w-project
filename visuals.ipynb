### Final Performance Comparison

```python
import numpy as np
import matplotlib.pyplot as plt

models = ['Baseline LightGCN', 'LightGCN+Attention']
final_rmse = [1.1716, 1.1186]
final_recall = [0.0800, 0.0531]

x = np.arange(len(models))
width = 0.35

fig, ax1 = plt.subplots(figsize=(8,5))
rmse_bars = ax1.bar(x - width/2, final_rmse, width, label='RMSE', color='skyblue')
ax1.set_ylabel('RMSE')
ax1.set_ylim(0,1.3)

ax2 = ax1.twinx()
recall_bars = ax2.bar(x + width/2, final_recall, width, label='Recall@10', color='salmon')
ax2.set_ylabel('Recall@10')
ax2.set_ylim(0,0.1)

ax1.set_xticks(x)
ax1.set_xticklabels(models)
ax1.set_title('Final Performance Comparison')

# Get handles and labels from both axes
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()

# Create a combined legend
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

fig.tight_layout()
plt.show()

```

### User-Item Interaction Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# Suppose we take a small subset of user-item pairs for demonstration
user_nodes = ['U1','U2','U3']
item_nodes = ['I1','I2','I3','I4','I5']
edges = [
    ('U1','I1'), ('U1','I2'),
    ('U2','I2'), ('U2','I3'),
    ('U3','I2'), ('U3','I5'),
    ('U1','I4'), ('U2','I4')
]

G = nx.Graph()
G.add_nodes_from(user_nodes, bipartite=0)
G.add_nodes_from(item_nodes, bipartite=1)
G.add_edges_from(edges)

pos = {}
pos.update((u, (0, i)) for i,u in enumerate(user_nodes))
pos.update((m, (1, j)) for j,m in enumerate(item_nodes))

plt.figure(figsize=(8,4))
nx.draw(G, pos, with_labels=True, node_color=['skyblue' if n in user_nodes else 'lightgreen' for n in G.nodes()])
plt.title('Sample User-Item Interaction Graph')
plt.show()
```
